# Simulación de Monte Carlo

monte_carlo_cover.jpg

La simulación de Montecarlo es un **método estadístico**. Este es utilizado para resolver problemas matemáticos complejos a través de la **generación de variables aleatorias**.

La simulación de Montecarlo, o método de Montecarlo, le debe el nombre al famoso casino del principado de Mónaco. La ruleta es el juego de casino más famoso y también el ejemplo más sencillo de mecanismo que permite generar números aleatorios.

La clave de este método está en entender el término **‘simulación’**. Realizar una simulación consiste en repetir, o duplicar, las características y comportamientos de un sistema real. Así pues, el objetivo principal de la simulación de Montecarlo es intentar imitar el comportamiento de variables reales para, en la medida de lo posible, analizar o predecir cómo van a evolucionar.

A través de la simulación, se pueden resolver desde problemas muy sencillos, hasta problemas muy complejos.

# Problema de Monty Hall

El Problema de Monty Hall es un problema de probabilidad que está inspirado por el concurso televisivo estadounidense Let's Make a Deal(Hagamos un trato). , famoso entre 1963 y 1986. Su nombre proviene del presentador, **Monty Hall**.

En este concurso, el concursante **escoge una puerta entre tres**, y su premio consiste en lo que se encuentra detrás. Una de ellas oculta un coche, y tras las otras dos hay una cabra. Sin embargo, antes de abrirla, el presentador, que sabe donde esta el premio, abre una de las otras dos puertas y muestra que detrás de ella hay una cabra. Ahora tiene el concursante una última oportunidad de cambiar la puerta escogida **¿Debe el concursante mantener su elección original o escoger la otra puerta? ¿Hay alguna diferencia?**

Ahora suponga que el concursante elige la **puerta 1.** A continuación, el anfitrión abre la **tercera** puerta, que tiene una cabra. A continuación, el presentador le pregunta al concursante si quiere cambiar las puertas.

Veremos por qué es más ventajoso cambiar la puerta:

podemos ver que después de que el anfitrión abre la puerta 3, la probabilidad de que las dos últimas puertas tengan un automóvil aumenta a 2/3. Ahora que sabemos que la tercera puerta tiene una cabra, la probabilidad de que la segunda puerta tenga un automóvil aumenta a 2/3. Por tanto, es más ventajoso cambiar las puertas.

Ahora vamos a utilizar el método de Monte Carlo para realizar este caso de prueba muchas veces y averiguar sus probabilidades de forma experimental.

```python
import random
import matplotlib.pyplot as plt 
```

```python
# Que hay en las 3 puertas 

puertas = ['Cabras', 'Cabras', 'Carro']

#Lista vacia para almacenar los valores de probabilidad
ganarOtraPuerta = []
ganarPuertaOriginal = []

plt.axhline(y = 0.66666, color = 'r', linestyle = '-')
plt.axhline(y = 0.33333, color = 'g', linestyle = '-')

def monte_carlo(vecesAccion):
    otraPuerta = 0
    puertaOriginal = 0

    for accion in range(vecesAccion):
        # Colocar aleatoriamente el coche y las cabras detrás de las tres puertas.
        random.shuffle(puertas)

        # Puerta elegida por el participante
        # Nota solo son 2: 0,1 las probabilidades por que el presentador descarta una puerta
        eleccion = random.randrange(2)

        # Participante no gana el carro
        if puertas[eleccion] != 'Carro':
            otraPuerta += 1
        else:
            puertaOriginal += 1
        
        # Ingresar los resultados obtenidos
        ganarOtraPuerta.append(otraPuerta/(accion+1))
        ganarPuertaOriginal.append(puertaOriginal / (accion+1))

        # Graficar resultados

        plt.plot(ganarOtraPuerta)
        plt.plot(ganarPuertaOriginal)
    
    # Explicación escrita

    print('Ganar si tu cambias la puerta', ganarOtraPuerta[-1])
    print('Ganar si no cambias la puerta', ganarPuertaOriginal[-1])

# Llamar la función Monte Carlo

monte_carlo(1000)
            
            




```

mostramos que después de 1000 iteraciones, la probabilidad de ganar si cambiamos la puerta es 0.66. Por lo tanto, estamos seguros de que nos conviene cambiar la puerta en este ejemplo.
